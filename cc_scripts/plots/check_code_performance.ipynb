```python
import json
import numpy as np
import os

from tqdm.notebook import tqdm
```

```python
rollout_dir = "/home/chanb/scratch/hint_rl_results/logs/chanb/eval-code/local_eval-evaluate_code-50_hints-openreasoning-nemotron/rollout/0"
```

```python
results = dict()

for filename in tqdm(os.listdir(rollout_dir)):
    full_file_path = os.path.join(rollout_dir, filename)

    if not os.path.isfile(full_file_path):
        continue

    key = int(filename.split(".")[0])
    curr_results = []
    for line in open(full_file_path):
        curr_sample = json.loads(line)
        curr_results.append(curr_sample["reward"])

    results[key] = (curr_results, curr_sample["question_id"])
```

```python
accept_range = [0, 4]
filtered_results = dict()
for key, vals in tqdm(results.items()):
    if accept_range[0] <= np.sum(vals[0]) <= accept_range[1]:
        filtered_results[key] = vals
```
